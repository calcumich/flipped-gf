```python
import re
import pandas as pd

def parse_entry(entry):
    try:
        if not isinstance(entry, str):
            print(entry)
            raise ValueError("Entry is not a string")

        task_description_match = re.search(r'### Task:\n(.*?)\n###', entry, re.DOTALL)
        task_description = task_description_match.group(1).strip() if task_description_match else "No Task Description"

        question_match = re.search(r'### Question:\n(.*?)\n###', entry, re.DOTALL)
        question = question_match.group(1).strip() if question_match else "No Question"

        captions_match = re.search(r'### Captions \(ID, Caption\):\n(.*?)\n###', entry, re.DOTALL)
        captions_text = captions_match.group(1).strip() if captions_match else "No Captions"
        captions = re.findall(r'(\d+): (.*?)\n', captions_text)

        instructions_match = re.search(r'### Instructions:\n(.*?)\nOutput:', entry, re.DOTALL)
        instructions = instructions_match.group(1).strip() if instructions_match else "No Instructions"

        output_match = re.search(r'Output: (\d+), (\d+).', entry)
        useful_ids = {int(output_match.group(1)), int(output_match.group(2))} if output_match else set()

        data = []
        for cap_id, cap_text in captions:
            data.append({
                "Task Description": task_description,
                "Question": question,
                "Caption ID": int(cap_id),
                "Caption Text": cap_text,
                "Useful for Answering": "Yes" if int(cap_id) in useful_ids else "No",
                "Additional Notes": instructions
            })
        
        return pd.DataFrame(data)

    except Exception as e:
        print(f"Failed to parse entry: {e}")
        return pd.DataFrame()  

def process_data(data_list):
    all_data = pd.DataFrame()  

    for entry in data_list:
        entry_data = parse_entry(entry)
        all_data = pd.concat([all_data, entry_data], ignore_index=True)

    return all_data

```

```python
import re

def parse_data(data):
    items = re.split(r'</s>\s*<s>', data.strip('<s>').strip('</s>'))

    parsed_items = []

    for item in items:
        parsed_item = {}

        task_match = re.search(r'### Task:\n(.*?)\n\n###', item, re.DOTALL)
        if task_match:
            parsed_item['Task'] = task_match.group(1).strip()

        question_match = re.search(r'### Question:\n(.*?)\n\n###', item, re.DOTALL)
        if question_match:
            parsed_item['Question'] = question_match.group(1).strip()

        captions_match = re.search(r'### Captions \(ID, Caption\)\):\n(.*?)\n\n###', item, re.DOTALL)
        if captions_match:
            captions = {}
            caption_lines = captions_match.group(1).split('\n')
            for line in caption_lines:
                caption_id, caption_text = line.split(':', 1)
                captions[caption_id.strip()] = caption_text.strip()
            parsed_item['Captions'] = captions

        instructions_match = re.search(r'### Instructions:\n(.*?)\n\n###', item, re.DOTALL)
        if instructions_match:
            parsed_item['Instructions'] = instructions_match.group(1).strip()

        answer_match = re.search(r'### Answer:\n(.*?)(\n\n|$)', item, re.DOTALL)
        if answer_match:
            parsed_item['Answer'] = answer_match.group(1).strip()

        parsed_items.append(parsed_item)

    return parsed_items

```
